```python
import torch
from allennlp.modules.elmo import Elmo, batch_to_ids
print(torch.__version__)

# Load ELMo model options and weights
options_file = "https://allennlp.s3.amazonaws.com/models/elmo/2x4096_512_2048cnn_2xhighway/elmo_2x4096_512_2048cnn_2xhighway_options.json"
weight_file = "https://allennlp.s3.amazonaws.com/models/elmo/2x4096_512_2048cnn_2xhighway/elmo_2x4096_512_2048cnn_2xhighway_weights.hdf5"

# Create an ELMo instance
elmo = Elmo(options_file, weight_file, num_output_representations=1, dropout=0)

# Define input sentences
sentences = [
    ["I", "have", "a", "cat"],
    ["Somebody", "just", "told", "me", "a", "joke"]
]

# Convert sentences to character ids and create a tensor
character_ids = batch_to_ids(sentences)

# Get ELMo embeddings
embeddings = elmo(character_ids)

# Extract ELMo embeddings for each token
elmo_embeddings = embeddings["elmo_representations"][0]

print("ELMo embeddings shape:", elmo_embeddings.shape)
print(elmo_embeddings)
```
